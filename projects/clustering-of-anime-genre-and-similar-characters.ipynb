```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
```

# <a id='tfidf'> PreProcessing</a>

```python
quotes = pd.read_csv("../input/anime-quotes/AnimeQuotes.csv")
quotes
```

```python
anime_groups = quotes.groupby('Anime').agg(lambda x: ' '.join(x))
character_groups=  quotes.groupby('Character').agg(lambda x:' '.join(x))
```

```python
tfidf_anime = TfidfVectorizer(ngram_range=(1,5), min_df=2)
quote_tfidf_anime = tfidf_anime.fit_transform(anime_groups.Quote)

tfidf_character = TfidfVectorizer(ngram_range=(1,5), min_df=2)
quote_tfidf_character = tfidf_character.fit_transform(character_groups.Quote)
```

```python
inertia = {}
for i in range(2, 20):
    kmeans = KMeans(n_clusters=i).fit(quote_tfidf_anime)
    inertia[i] = kmeans.inertia_
```

```python
plt.scatter(x=inertia.keys(), y=inertia.values())
```

There doesn't seem to be that much of a difference in inertia across the cluster numbers. So I chose 10 as the ideal number of clusters without any specific reason.

```python
kmeans = KMeans(n_clusters=10)
clusters = kmeans.fit_predict(quote_tfidf_anime)
```

```python
anime_groups['cluster'] = clusters
```

```python
anime_groups = anime_groups.reset_index()[['Anime', 'Quote', 'cluster']]
anime_groups
```

# <a id='anime'> Anime Grouping </a>

```python
for index, row in anime_groups.groupby(by='cluster'):
    print("Cluster {}".format(index+1))
    print(row['Anime'].values) 
```

```python
inertia = {}
for i in range(2, 20):
    kmeans = KMeans(n_clusters=i).fit(quote_tfidf_character)
    inertia[i] = kmeans.inertia_
```

```python
plt.scatter(x=inertia.keys(), y=inertia.values())
```

```python
kmeans = KMeans(n_clusters=7)
clusters = kmeans.fit_predict(quote_tfidf_character)
```

```python
character_groups['cluster'] = clusters
```

```python
character_groups = character_groups.reset_index()[['Character', 'Quote', 'cluster']]
character_groups
```

# <a id='characters'> Character Grouping </a>

```python
for index, row in character_groups.groupby(by='cluster'):
    print("Cluster {}".format(index+1))
    print(row['Character'].values) 
```
